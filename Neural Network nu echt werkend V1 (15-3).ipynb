# Initialization

Here we initialize the imports

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
from keras.optimizers import SGD, Adam, rmsprop
from sklearn.decomposition import PCA
import pandas as pd
from IPython.display import SVG
from keras.utils import plot_model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
#from pydot import *
#from graphviz import *

dataset = pd.read_csv('./diabetes.csv', delimiter=",")

```

```python
split = 500

values = dataset.values[:,0:8]

pca = PCA(n_components = 8)
pca.fit(values)

values = pca.transform(values)
#Xtrain = pca.transform(Xtrain)
#Xtest = pca.transform(Xtest)

#print (values)

Xtrain = values[:split,0:8]
Ytrain = dataset.values[:split,8]
Xtest = values[split:,0:8]
Ytest = dataset.values[split:,8]


#model = Sequential()
#model.add(Dense(16, input_dim=8, activation='relu'))
#model.add(Dense(30, activation='relu'))
#model.add(Dense(10, activation='selu'))
#model.add(Dense(8, activation='relu'))
#model.add(Dense(10, activation='selu'))
#model.add(Dense(1, activation='sigmoid'))

model = Sequential()
model.add(Dense(100, input_dim=8, activation = 'relu'))
model.add(Dense(50, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))


optimizer = rmsprop(lr=0.001)

model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

model.fit(Xtrain, Ytrain, epochs=30, batch_size=10)
model.fit(Xtrain, Ytrain, epochs=5, batch_size=32, validation_split=1/6)

# calculate predictions
predictions = [round(x[0]) for x in model.predict(Xtest)]
#print (len(predictions))
#print (len(Ytest))
print(confusion_matrix(Ytest, predictions))
print("Accuracy is ", accuracy_score(Ytest,predictions)*100)

tn, fp, fn, tp = confusion_matrix(Ytest, predictions).ravel()
print('TN FP FN TP')
print((tn, fp, fn, tp))
print ('Precision is =', tp/(tp+fp))
print ('Recall is =', tp/(tp + tn))


```

```python
TP = 0
TN = 0
FP = 0
FN = 0

for i in range(len(predictions)):
    if predictions[i] == Ytest[i]:
        if predictions[1] == 1:
            TP += 1
        else:
            TN += 1
    else:
        if predictions[i] == 1:
            FP += 1
        else:
            FN += 1
            
#plot_model(model, to_file='model.png')           

print('TP TN FP FN')
print(TP, TN, FP, FN)            
            
print ('Precision is =', TP/(TP+FP))
print ('Recall is =', TP/(TP + TN))
print ('Accuracy is =', (TP+TN)/(TP+TN+FP+FN))
```
